```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import scipy as sy
import pylab as plb  
import pandas as pd
import numpy as np
```

```python
dt = pd.read_csv('/home/jovyan/work/selectShop.csv', index_col=0,encoding='cp950') 
#dt=dt+0.0000001
dt  #欄位K11望小,,欄位K12望目,其他都是望大
```

```python
dt['K11']=1/dt.K11  #望小::原數值改為倒數
dt
```

```python
#望目 =1-[ABS(x-OB)/MAX]
OB=2
dt['K12']=1-(abs(dt.K12-OB)/max(dt.K12))
dt
```

```python
cmax=dt.max(axis=0)  #直行
cmin=dt.min(axis=0)  #直行

x0=dt

#轉為倍數B=cmax/ x0
B= cmax/ x0

#關聯度公式G= (Bmin+Bmax)/(B+ Bmax)

#Bmax=B.max(axis=0)  #直行
Bmax=B.values.max() 
#Bmin=B.min(axis=0)  #直行
Bmin= B.values.min()  #區塊

G= (Bmin+Bmax)/(B+ Bmax)

#關聯度(橫列平均)
Gavg=np.mean(G,axis=1)

ngama=-np.array(Gavg)
rank= ngama.argsort()
rank=rank.argsort()

names = dt.columns.values 
x0 = pd.DataFrame(data=x0,columns=names) 
x0= round(x0,3)
x0['avg'] = Gavg
x0['avg'] = round(x0['avg'],3)
x0['rank'] = rank+1
x0
```
